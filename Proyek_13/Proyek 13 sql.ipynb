 Identifikasi penerbit yang telah menerbitkan jumlah buku terbanyak, dengan jumlah halaman lebih dari 50 (hal ini akan membantu Anda dalam mengecualikan brosur dan publikasi serupa dari analisis Anda).

```python
query="SELECT p.publisher, COUNT(*) AS Jumlah FROM books b INNER JOIN publishers p ON p.publisher_id = b.publisher_id WHERE b.num_pages > 50 GROUP BY p.publisher ORDER BY Jumlah DESC"

publisher_book = pd.io.sql.read_sql(query, con=engine)
publisher_book
```

Identifikasi penulis dengan rating rata-rata buku tertinggi: temukan buku dengan rating minimal 50

```python
query="SELECT a.author,b.title, COUNT(r.rating)FROM books b INNER JOIN authors a ON b.author_id = a.author_id INNER JOIN ratings r ON b.book_id = r.book_id GROUP BY a.author,b.title HAVING COUNT(r.rating )>50 "

book_rating = pd.io.sql.read_sql(query, con=engine)
book_rating
```

Hitung jumlah rata-rata teks ulasan di antara pengguna yang memberi rating terhadap lebih dari 50 buku

```python
query = " SELECT re.username, AVG(LENGTH(text)) AS panjang_ulasan FROM books b INNER JOIN reviews re ON b.book_id = re.book_id INNER JOIN ratings r ON b.book_id = r.book_id WHERE LENGTH(text) >50 IS NOT NULL GROUP BY re.username UNION ALL SELECT 'TOTAL AVG ',AVG(LENGTH(text)) from reviews WHERE LENGTH(text)>50 "
long = pd.io.sql.read_sql(query,con=engine)
long
```
